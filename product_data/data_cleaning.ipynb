## Data Cleaning of Amazon Product Data

```python
import pandas as pd
```

```python
df = pd.read_csv('amazon_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df['actual_price'] = (
    df['actual_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df['discounted_price'] = (
    df['discounted_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df.head()
```

```python
df = df.to_csv('amazon_data.csv', index=False)
```
